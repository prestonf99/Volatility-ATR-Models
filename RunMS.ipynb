**_Available Tickers_**

```python
sp = ms.Vols(360, '^GSPC')  # (days_to_load, S&P)
nq = ms.Vols(360, '^NDX')   # Nasdaq 100
rut = ms.Vols(360, '^RUT')  # Russell 2000
dji = ms.Vols(360, '^DJI')  # Dow Jones
cl = ms.Vols(360, 'CL=F')   # Crude Oil
gld = ms.Vols(360, 'GLD')   # Gold
aapl = ms.Vols(360, 'AAPL') # Apple
amzn = ms.Vols(360, 'AMZN') # Amazon
eem = ms.Vols(360, 'EEM')   # Emerging Markets
gs = ms.Vols(360, 'GS')     # Goldman Sachs
goog = ms.Vols(360, 'GOOG') # Google
```
