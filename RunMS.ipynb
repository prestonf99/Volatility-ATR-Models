```python
import MorningStats as ms
```

You can adjust the days to load, the ATR will always come out the same way, but if you want to see a longer time series of the VRP, just increase the first number. It currently only supports S&P(^GSPC) & Nasdaq. 

```python
sp = ms.MorningStats(360, '^GSPC') #(days_to_load, symbol)
nq = ms.MorningStats(360, '^NDX') 
```

### Average Trading Range

```python
sp.atr()
```

```python
nq.atr()
```
